```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
face1 = cv2.imread('./homework-facedata/1.jpg')
face2 = cv2.imread('./homework-facedata/2.jpg')
face3 = cv2.imread('./homework-facedata/3.jpg')
face1 = cv2.cvtColor(face1, cv2.COLOR_BGR2HSV)
face2 = cv2.cvtColor(face2, cv2.COLOR_BGR2HSV)
face3 = cv2.cvtColor(face3, cv2.COLOR_BGR2HSV)
```

```python
def eda(img_list, channel):
    d = {}
    for img in img_list:
        for i in range(img.shape[0]):
            for j in range(img.shape[1]):
                if img[i, j, channel] not in d:
                    d[img[i, j, channel]] = 0
                d[img[i, j, channel]] += 1
    d = dict(sorted(d.items()))
    plt.plot(d.keys(), d.values())
```

```python
face_eda = [face1[270:1100, 650:1150, :], face2[150:450,400:570, :], face3[120:310, 300:430, :], face3[80:310, 800:930, :]]
```

```python
eda(face_eda, 0)
```

```python
eda(face_eda, 1)
```

```python
eda(face_eda, 2)
```

```python
def create_result(face, lower, upper, output_name):
    mask = cv2.inRange(face, lower, upper)
    res = cv2.bitwise_and(face, face, mask = mask)
    res = cv2.cvtColor(res, cv2.COLOR_HSV2BGR)
    cv2.imwrite(output_name+'.jpg', res)
    plt.imshow(res[:, :, ::-1])
```

```python
lower = np.array([0, 50, 160])
upper = np.array([13, 150, 255])
```

```python
create_result(face1, lower, upper, '1')
```

```python
create_result(face2, lower, upper, '2')
```

```python
create_result(face3, lower, upper, '3')
```
