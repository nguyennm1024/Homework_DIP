```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def export_result(ori_image, mask, result_name):
    contours, hierarchy = cv2.findContours(mask, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    ori_image = cv2.drawContours(ori_image, contours, 0, (0,0,255), 3)
    plt.imshow(ori_image[:,:,::-1])
    cv2.imwrite(result_name+".jpg", ori_image)
```

```python
image = cv2.imread("homework-data/01.jpg", 0)
ret, thresh = cv2.threshold(image, 150, 255, cv2.THRESH_BINARY)
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))
opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel)
dilation = cv2.dilate(opening, kernel, iterations=100)
dilation = cv2.erode(dilation, kernel, iterations=100)
plt.imshow(dilation, cmap='gray')
```

```python
nlabel,labels,stats,centroids = cv2.connectedComponentsWithStats(dilation,connectivity=8)
for i in range(1, nlabel):
    print(stats[i, cv2.CC_STAT_HEIGHT])
```

```python
mask = np.zeros_like(dilation, dtype=np.uint8)
mask[labels==4] = 255
plt.imshow(mask, cmap='gray')
```

```python
ori_image = cv2.imread("homework-data/01.jpg")
export_result(ori_image, mask, '01_contour')
```
